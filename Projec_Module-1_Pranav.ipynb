```python
#Taking the txt file from pratik and performing further operation like removing stop words, stemming and tagging
text_path = 'modified_file.txt'

with open(text_path, 'r') as file:
    text = file.read()

text
```

```python
#word tokenization of txt file
word = word_tokenize(text)
print(word)
```

```python
#sentence tokenization of txt file
sent = sent_tokenize(text)
for i in sent:
    print(i)
```

```python
#Removing stop words and printing filtered word  
filtered_word = [ w for w in  word if w not in stop_words]
print(filtered_word)
```

```python
print("length of original word in text ",len(word))
print("length of word after removing stop words",len(filtered_word))
```

```python
#Applying Porter stemming after filtering the stop words
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed_word = []
for w in filtered_word:
    stem_word = ps.stem(w)
    stemmed_word.append(stem_word)
print(stemmed_word)
```

```python
# Performing part-of-speech tagging in whole words
tagged_tokens = nltk.pos_tag(stemmed_word)

# Print the result
print(tagged_tokens)
```

```python
# Performing part-of-speech tagging in only noun words

# Filter only nouns (NN, NNS, NNPS, and NNP)
Tag_n_tks = [(word, pos) for word, pos in tagged_tokens if pos.startswith('NN')]

# Print the tagged noun tokens
print("Tagged Noun Tokens:", Tag_n_tks)

```
